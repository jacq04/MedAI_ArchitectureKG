## PHASE1: DATA CLEANING

```python
import pandas as pd

raw_data = "RawData.xlsx"
sheets = pd.ExcelFile(raw_data).sheet_names

print("Sheets found:", sheets)
```

```python
#STEP 1 - REMOVE FIRST ROW
df1 = [] #list per sheet but with the top row deleted
for sheet in sheets:
    df = pd.read_excel(raw_data, sheet_name=sheet, skiprows = 1) 
    df1.append(df)
df1[0].head()
```

```python
#STEP 2 - COMBINING ALL THE SHEETS
combined_df = pd.concat(df1, ignore_index=True)

#combined_df.head()
```

```python
#STEP3 - DROPPING ROWS WITHOUT A BOXOLOGY
filtered_df = combined_df[
    combined_df["Description and Clinical application"].notna() &
    (combined_df["Description and Clinical application"].astype(str).str.strip() != "")
]
print("Number of remaining rows:", filtered_df.shape[0])
filtered_df.head()
```

```python
#STEP4 - ASSIGNING INDEXS
filtered_df = filtered_df.reset_index(drop=True)  
filtered_df.insert(0, "Index", filtered_df.index + 1) 
filtered_df.head()
```

```python
#exporting data
filtered_df.to_excel("ProcessedData.xlsx", index=False)

print("DATA EXPORTED")
```
